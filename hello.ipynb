## Étape 1 - Validation d'une saisie

```java
import java.util.Scanner;

// Préparation de l'outil pour lire le clavier
Scanner clavier = new Scanner(System.in);
int nombre;

// La boucle commence ici
do {
    System.out.println("Veuillez saisir un nombre entre 2 et 15 : ");
    // On attend que l'utilisateur tape un nombre
    nombre = clavier.nextInt();

    // Si le nombre n'est pas valide, on affiche le message d'erreur
    if (nombre < 2 || nombre > 15) {
        System.out.println("Erreur : " + nombre + " n'est pas valide.");
    }

} while (nombre < 2 || nombre > 15); // On recommence TANT QUE c'est invalide

// Si on sort de la boucle, c'est que c'est gagné
System.out.println("Saisie validée : " + nombre);
```

## Étape 2 - Dessin de ligne(s)

```java

// Étape 2 : Carré creux ou plein
System.out.print("Voulez-vous un carré creux ? (oui/non) : ");
String choix = clavier.next();

for (int i = 0; i < nombre; i++) {
    for (int j = 0; j < nombre; j++) {
        // Si première/dernière ligne OU première/dernière colonne
        if (choix.equals("oui") && i > 0 && i < nombre - 1 && j > 0 && j < nombre - 1) {
            System.out.print("  "); // Espace vide au milieu
        } else {
            System.out.print("* ");
        }
    }
    System.out.println();
}
```

## Étape 3 - Dessin de triangle(s)

```java
import java.util.Scanner;

Scanner clavier = new Scanner(System.in);
int nombre;

// Étape 1 : Validation
do {
    System.out.println("Veuillez saisir un nombre entre 2 et 15 : ");
    nombre = clavier.nextInt();
    
    if (nombre < 2 || nombre > 15) {
        System.out.println("Erreur : " + nombre + " n'est pas valide.");
    }
} while (nombre < 2 || nombre > 15);

System.out.println("Saisie validée : " + nombre);
System.out.println();


// Étape 2 : Sapin
for (int i = 1; i <= nombre; i++) {
    for (int j = 0; j < nombre - i; j++) {
        System.out.print(" ");
    }
    for (int j = 0; j < (2 * i - 1); j++) {
        System.out.print("*");
    }
    System.out.println();
}
```

## Etape 4 : Coeur

```java
// Coeur dynamique basé sur le nombre saisi
System.out.println();

int n = nombre;

// Partie haute du coeur (les deux bosses)
for (int i = n/2; i <= n; i += 2) {
    // Espaces avant
    for (int j = 1; j < n - i; j += 2) {
        System.out.print(" ");
    }
    // Première bosse
    for (int j = 1; j <= i; j++) {
        System.out.print("*");
    }
    // Espaces au milieu
    for (int j = 1; j <= n - i; j++) {
        System.out.print(" ");
    }
    // Deuxième bosse
    for (int j = 1; j <= i; j++) {
        System.out.print("*");
    }
    System.out.println();
}

// Partie basse du coeur (pointe)
for (int i = n; i >= 1; i--) {
    // Espaces avant
    for (int j = i; j < n; j++) {
        System.out.print(" ");
    }
    // Étoiles
    for (int j = 1; j <= (i * 2) - 1; j++) {
        System.out.print("*");
    }
    System.out.println();
}
```
